```python
import numpy as np
import pandas as pd
from get_map_data import GetMapData
import json
import math
```

```python
f1 = open('./data/tokyo_hotels.json', "r")
data1 = json.loads(f1.read())
f1.close()

f2 = open('./data/london_hotels.json', "r")
data2 = json.loads(f2.read())
f2.close()

f3 = open('./data/singapore_hotels.json', "r")
data3 = json.loads(f3.read())
f3.close()

f4 = open('./data/new_york_hotels.json', "r")
data4 = json.loads(f4.read())
f4.close()
```

```python
df = None


def add_city(json_data, city):
    global df
    df_new = pd.json_normalize(json_data)
    df_new['city'] = city
    if df is None:
        df = df_new
    else:
        df = df.append(df_new, ignore_index=True)
    return df
```

```python
add_city(data1, 'Tokyo')
add_city(data2, 'London')
add_city(data3, 'Singapore')
add_city(data4, 'New York')
```

```python
location_df = df[['city','lat_long_link']]
location_df.head(10)
```

```python
location_df.shape
```

```python
get_map_data = GetMapData()
```

```python
lat = []
lon = []

for index, row in location_df.iterrows():
    coordinates = get_map_data.get_location(location=row['lat_long_link'])
    if coordinates is not None:
        lat.append(coordinates[0])
        lon.append(coordinates[1])
        print(index,"    ----> ",coordinates)
    else:
        lat.append(np.nan)
        lon.append(np.nan)
        print(index)
```

```python
location_df['lat'] = lat
```

```python
location_df['lon'] = lon
```

```python
location_df.info()
```

```python
location_df = location_df.drop(['lat_long_link'],axis=1)
location_df.info()
```
